```python
from collections import deque
import sys

def bfs_shortest_path(graph, start, goal):
    queue = deque([[start]])
    visited = set([start])

    print(" Searching shortest path:-\n")

    while queue:
        path = queue.popleft()
        node = path[-1]


        sys.stdout.write("Exploring: " + " -> ".join(path) + "\n")

        if node == goal:
            print("\n Goal reached!")
            return path

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                new_path = path + [neighbor]
                queue.append(new_path)

    return None
```

```python
graph = {}

n=int(input("Enter number of nodes: "))

for i in range(n):
    node = input("Enter node: ").upper()
    neighbours = input("Enter neighbours of {} separated by space: ".format(node)).upper().split()
    graph[node] = neighbours

s=input("Enter the root node:").upper()
g=input("Enter the goal Node:").upper()

path = bfs_shortest_path(graph, s, g)

if path:
    print("\n Shortest Path Found:")
    print("  " + " → ".join(path))
    print(" Distance:", len(path) - 1)
else:
    print("No path found")
```
